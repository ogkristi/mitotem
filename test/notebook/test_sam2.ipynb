### Visual tests for sam2

```python
from mitotem.predictor import Sam2Predictor
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread("data/interim/mito.tif", cv.IMREAD_GRAYSCALE)
sam = Sam2Predictor()
mask = sam.predict(img)
```

```python
fig, axes = plt.subplots(1,2, figsize=(12,6))
for ax, x in zip(axes,(img,mask)):
    ax.imshow(x, cmap='gray')
    ax.set_axis_off()
```
